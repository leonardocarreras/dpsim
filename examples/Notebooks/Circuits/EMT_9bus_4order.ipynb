```python
import matplotlib.pyplot as plt
%matplotlib widget

### NO LOADSTEP ###

gen_vars = {
    'i_gen1': ['GEN1.I_0', 'GEN1.I_1', 'GEN1.I_2'],
    'i_gen2': ['GEN2.I_0', 'GEN2.I_1', 'GEN2.I_2'],
    'i_gen3': ['GEN3.I_0', 'GEN3.I_1', 'GEN3.I_2'],
}


dataframes = [df1, df2, df3, df4, df5]
titles = ['Sim']

fig, axes = plt.subplots(1, 3, figsize=(12, 3))  # 1 row, 3 columns

df = dataframes[0]  # Since you only have one dataframe

for ax, (gen_name, var_list) in zip(axes, gen_vars.items()):
    for var in var_list:
        ax.plot(df['time'].values, df[var].values, linestyle="--", label=var.split('_')[-1])
    ax.set_title(gen_name)
    ax.set_xlabel('Time')
    ax.set_ylabel('Current')
    ax.grid(True)
    ax.legend(title="Phase")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

dataframes = [df1, df2, df3, df4, df5]
phases = ['BUS6_0', 'BUS6_1', 'BUS6_2']
num_runs = len(dataframes)

fig, axes = plt.subplots(nrows=num_runs, ncols=len(phases), 
                         figsize=(15, 3*num_runs), sharex=True)

for i, df in enumerate(dataframes):
    for j, phase in enumerate(phases):
        ax = axes[i, j]
        ax.plot(df['time'], df[phases])
        ax.set_title(f'Run {i+1} – {phases}')
        ax.set_ylabel('Voltage [V]')
        if i == num_runs - 1:
            ax.set_xlabel('Time [s]')

plt.tight_layout()
plt.show()

### RUN 1 - no loadstep, RUN 2 - 33% Load, RUN 3 - 66% Load, RUN 4 - 133% Load, RUN 5 - 166% Load ###
```

```python
import matplotlib.pyplot as plt


dfs = [df1, df2, df3, df4, df5]
generators = {
    'GEN1': ['GEN1.I_0', 'GEN1.I_1', 'GEN1.I_2'],
    'GEN2': ['GEN2.I_0', 'GEN2.I_1', 'GEN2.I_2'],
    'GEN3': ['GEN3.I_0', 'GEN3.I_1', 'GEN3.I_2'],
}

num_runs = len(dfs)
num_gens = len(generators)

fig, axes = plt.subplots(nrows=num_runs, ncols=num_gens, 
                         figsize=(5*num_gens, 3*num_runs), sharex=True)

for i, df in enumerate(dfs):
    for j, (gen_name, cols) in enumerate(generators.items()):
        ax = axes[i, j]
        for col in cols:
            ax.plot(df['time'], df[col], label=col.split('.')[-1])
        ax.set_title(f'Run {i+1} – {gen_name}')
        ax.set_ylabel('Current [A]')
        if i == num_runs - 1:
            ax.set_xlabel('Time [s]')
        if i == 0:
            ax.legend()

plt.tight_layout()
plt.show()


### RUN 1 - no loadstep, RUN 2 - 33% Load, RUN 3 - 66% Load, RUN 4 - 133% Load, RUN 5 - 166% Load ###
```
